# Question 2:
How to create a series from a numpy array?

```python
import numpy as np

arr = np.array([1,2,3,4,5,6])
print(arr)
print(type(arr))

series = pd.Series(arr)
print(series)
print(type(series))
```

# Question 3:
How to convert the index of a series into a column of a dataframe?

```python
population_dict = {
    "California":38335215,
    "Texas":19665634,
    "New york":16545697,
    "Florida":78954632,
    "Illinois":12354698
}
series = pd.Series(population_dict)
print(series)
print("\nIndex of Series:", series.index)

df = pd.DataFrame(series)
df.reset_index(level = 0, inplace = True)
df.rename(columns = {"index": "City", 0: "Population"}, inplace = True)
df
```
